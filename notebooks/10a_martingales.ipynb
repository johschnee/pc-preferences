```python
# Simulation parameters
num_trials = 10000  # Number of independent simulations
max_n = 5000        # Maximum number of steps
print_each = 100

a = np.array([1, 0, 0])
eps = 0.1
rng = np.random.default_rng(1331)
stoch_process = np.zeros((num_trials, 3))

for m in range(2, max_n + 1):
    step = rng.beta(m-1, 1, (num_trials, 3))
    U, V, W = step[:, 0], step[:, 1], step[:, 2]
    
    stoch_process[:, 0] = U - V + U * V * stoch_process[:, 0]
    stoch_process[:, 1] = V - W + V * W * stoch_process[:, 1]
    stoch_process[:, 2] = W - U + W * U * stoch_process[:, 2]

    if m % print_each == 0:
        var = np.sum(a**2)*(m-1)*(2*m+5)/(3*m*(m+1)**2) + 2*(a[0]*a[1] + a[1]*a[2] + a[2]*a[0])*(1-m)/(3*m**2)
        X = (stoch_process @ a) / np.sqrt(var)
        print(m, var, np.mean(X**2 * (np.abs(X) >= eps).astype(int)), np.mean(np.abs(X) > eps))
    
# martingales.shape, martingales[:10], 
# np.cov(stoch_process.T)
```
